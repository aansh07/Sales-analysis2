```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("all_data.csv")
```

```python
df.head()
```

```python
df_clean=df.dropna()
```

```python
df_clean.head()
```

```python
df_clean.dtypes
```

```python
df_clean["Order ID"]=pd.to_numeric(df_clean["Order ID"],errors="coerce")
df_clean["Quantity Ordered"]=pd.to_numeric(df_clean["Quantity Ordered"],errors="coerce")
df_clean["Price Each"]=pd.to_numeric(df_clean["Price Each"],errors="coerce")
```

```python
df_clean.head()
```

```python
df_clean.dtypes
```

```python
df_clean["Order ID"]=df_clean["Order ID"].fillna(0).astype(int)
df_clean["Quantity Ordered"]=df_clean["Quantity Ordered"].fillna(0).astype(int)
```

```python
df_clean.dtypes
```

```python
df_clean.head()
```

#SPILLTING THE ADDRESS.

```python
df_clean["Street"]=df_clean["Purchase Address"].str.split(",").str[0]
df_clean["Area"]=df_clean["Purchase Address"].str.split(",").str[1]
df_clean["Country"]=df_clean["Purchase Address"].str.split(",").str[2]
df_clean.head(2)
```

```python
df_clean.drop(["Purchase Address"],axis=1,inplace=True)
```

```python
df_clean.head()
```

```python
df_clean["Date_Of_Order"]=df_clean["Order Date"].str.split(" ").str[0]
df_clean["Time_Of_Order"]=df_clean["Order Date"].str.split(" ").str[1]
df_clean.head()
```

```python
df_clean.drop(["Order Date"],axis=1,inplace=True)
```

```python
df_clean.head()
```

```python
df_clean["Month"]=df_clean["Date_Of_Order"].str.split("/").str[0]
```

```python
df_clean.head()
```

```python
df_clean["Total_Sales"]=df_clean["Quantity Ordered"].astype(int)*df_clean["Price Each"].astype(float)
```

```python
df_clean.head()
```

           #Question 1: What was the best month for sales? How much was earned that month?

```python
df_wordf = df_clean.groupby(["Month"]).sum()
```

```python
months = range(0, 13)  # or range(1, 14) depending on your data
plt.bar(months, df_wordf['Total_Sales'])
plt.show()
    
```
